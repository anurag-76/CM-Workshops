```python
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1 and 2

y1 = (5-x)/3

y2 = (5+x)/2

y3 = 3-x

## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(0.5,0.6, '$x+3y = 5$', color='r')

ax.text(-3,0.5,'$x-2y = -5$', color='g')

ax.text(1.2,2, '$x+y = 3$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-2,6)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

From the graph, we can conclude that each of the three equations is represented by a straight line, and while every pair of lines intersects at a unique point (indicating that each pair of equations has a consistent solution), all three lines do not intersect at a single common point. This means that although the systems formed by two equations at a time are solvable, the system of all three equations together is inconsistent and has no solution that satisfies them simultaneously. Geometrically, the intersections highlight the pairwise solutions, but the absence of a single meeting point confirms that no common solution exists for all three equations.

```python
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1 and 2

y1 = (3-2*x)/3

y2 = (6-4*x)/6

y3 = (1-3*x)/-2

## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(-4, 3, '$2x+3y=3$', color='r')

ax.text(-4, 2.5, '$4x+6y=6$', color='g')

ax.text(2, 2, '$3x-2y=1$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-4,4)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))
ax.set_yticks(list(range(-4,5)))

##finding intersection point
A = np.array([[2, 3],
              [3, -2]])
B = np.array([3, 1])
m = np.linalg.solve(A, B)
print("Intersection point:", m)

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

```python
import numpy as np

# Generate random 5x5 matrix
A = np.random.randint(1, 10, (5, 5))
print("Matrix A:\n", A)

# Rank
rank_A = np.linalg.matrix_rank(A)
print("\nRank of A:", rank_A)

# Check invertibility
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

if det_A != 0:
    print("Matrix A is invertible.")
else:
    print("Matrix A is NOT invertible.")
```

```python
# Generate random 5x5 matrix and vector
A = np.random.randint(1, 10, (5, 5))
x = np.random.randint(1, 10, (5, 1))

# Compute b = A x
b = A @ x

print("Matrix A:\n", A)
print("\nVector x:\n", x)
print("\nVector b = A x:\n", b)

# Solve the linear system A x = b
x_solution = np.linalg.solve(A, b)
print("\nSolution x:\n", x_solution)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given points
x = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 7])
y = np.array([-7.6, 1, 4.8, 5, 2.8, -0.6, -4, -6.2, -6, 6.4])

# Step 1: Plot given points
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y, '--', color='blue', label='Connected Points')
plt.title("Task 4: Given Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()

# Step 2: Fit a polynomial (degree 3 for smooth curve)
coeffs = np.polyfit(x, y, 3)
poly = np.poly1d(coeffs)
print("Polynomial coefficients:", coeffs)

# Step 3: Plot polynomial fit
x_fit = np.linspace(min(x), max(x), 200)
y_fit = poly(x_fit)

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_fit, y_fit, color='green', label='Polynomial Fit')
plt.title("Polynomial Fit to Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Check fit
y_pred = poly(x)
print("Predicted values:", y_pred)
print("Original values:", y)
```
