**Problem 3: Given two compatible matrices, write a Python function that performs Matrix-Matrix
product. Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-
Matrix product.
Hint: Follow that above example and extend it.**

```python
import numpy as np

def dot_product(vec1, vec2):
    # flatten the input vectors
    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))

def matrix_matrix_product(matA, matB):
    m, n = matA.shape
    p, q = matB.shape

    assert n == p, "The matrices are not compatible for multiplication!"

    # create a ndarray to store the result
    result = np.zeros((m, q), dtype=float)

    # Loop to perform Matrix-Matrix product
    for i in range(m):
        for j in range(q):
            # dot product between row i of matA and column j of matB
            result[i, j] = dot_product(matA[i, :], matB[:, j])

    return result


if __name__ == "__main__":
    """ Part 3: Problem 3: Matrix-Matrix product """
    mat1 = np.full((3,3), 3, dtype=int)
    mat2 = np.full((3,2), 2, dtype=int)

    print("Matrix A = ")
    print(mat1)
    print("Matrix B = ")
    print(mat2)

    # Call the function matrix_matrix_product(mat1, mat2)
    result1 = matrix_matrix_product(mat1, mat2)
    print("Result (custom function) = ")
    print(result1)

    # Check with numpy in-built function for multiplication
    result2 = mat1 @ mat2
    print("Result (NumPy @ operator) = ")
    print(result2)
```

**Problem 4: Given a matrix, write a Python function that checks if it is a symmetric matrix.
Hint: Use the definition based on Transpose of Matrix.**

```python
import numpy as np

def is_symmetric_matrix(mat):
    # Check if matrix is square
    if mat.shape[0] != mat.shape[1]:
        return False

    # Compare matrix with its transpose
    return np.array_equal(mat, mat.T)


if __name__ == "__main__":
    """ Part 4: Problem 4: Symmetric Matrix check """
    mat1 = np.array([[1, 2, 3],
                     [2, 5, 6],
                     [3, 6, 9]])

    print("Matrix = ")
    print(mat1)

    if is_symmetric_matrix(mat1):
        print("The matrix is symmetric.")
    else:
        print("The matrix is NOT symmetric.")
```
