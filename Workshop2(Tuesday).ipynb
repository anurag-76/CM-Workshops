1. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
Sample List : [1,2,3,3,3,3,4,5] Unique List : [1, 2, 3, 4, 5]

```python
def unique_elements(input_list):
    # Convert the list to a set to remove duplicates, then back to a list
    return list(set(input_list))

sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
print(unique_elements(sample_list))
```

2.	Write a Python function that takes a number as a parameter and checks whether the number is prime or not.


```python
def check_prime(n):
    #initialize f variable to count factors
    f = 0
    for i in range(1, n+1):
        if n % i == 0:
            f += 1

    #check for negative numbers , 0 and 1 which are not prime
    if n <= 1:
        return("The number is not PRIME")
    #if the factor of that number is 2 then it is prime else composite
    if f == 2:
        return("The number is PRIME")
    else:
        return("The number is COMPOSITE")

number = int(input("Enter any number: "))
check_prime(number)
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
# Function to check if a number is prime
def check_prime(n):
    #initialize f variable to count factors
    f = 0
    for i in range(1, n+1):
        if n % i == 0:
            f += 1

    #check for negative numbers , 0 and 1 which are not prime
    if n <= 1:
        return(False)
    #if the factor of that number is 2 then it is prime else composite
    if f == 2:
        return(True)
    else:
        return(False)

# Print twin primes less than 1000
# check only odd numbers
for i in range(3, 1000, 2):
    if check_prime(i) and check_prime(i + 2):
        print(f"({i}, {i+2})")
```

4. positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):
    total = 0
    #run a loop to find the factors of number excluding itself
    for i in range(1, n):
        if n % i == 0:
            total += i
    #check if sum of divisor is equal to number itself
    if (total == n):
        print("It is perfect number")
    else:
        print("It is not a perfect number")

sumOfDivisors(6)
sumOfDivisors(28)
sumOfDivisors(255)
```

```python
def SumOfDivisors(n):
    total = 0
    #run a loop to find the factors of number excluding itself
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

def NextPerfectNumber(num):
    while True:
        #Check If num is perfect number
        if SumOfDivisors(num) == num:
            return num
        num += 1

# The smallest perfect number is 6. We start checking from 7.
print(f"The next perfect number after 6 is: {NextPerfectNumber(7)}")
```

6.Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
# Below is given list according to question
Marks = [20, 30, 35, 55, 70, 80, 90]

#calculating Mean
def calc_mean(data):
  sum = 0
  for i in data:
    sum += i
  return (sum / len(data))

#calculating Median
def calc_median(data):
  count = len(data)           #calculates length of list
  data.sort()                 #arranges the elements inside the list in ascending order
  if(count % 2 == 1):         #Odd number of Elements
    return data[count // 2]
  else:                       #Even number of Elements
    m1 = data[(count // 2)-1]
    m2 = data[count // 2]
    return (m1 + m2) / 2      #average of two mid values

print("The Mean of the given list : ",calc_mean(Marks))
print("The Median of the given list : ",calc_median(Marks))
```

7. Write a function that converts a decimal number to binary number.

```python
def DecimalToBinary(n):
  binary = ""
  while n > 0:
    rem = n%2
    binary = str(rem) + binary
    n = n // 2
  if (binary == ""):
    return "0"
  else:
    return binary

print("The Binary of Decimal 10 is : "+DecimalToBinary(10))
print("The Binary of Decimal 2 is : "+DecimalToBinary(2))
print("The Binary of Decimal 25 is : "+DecimalToBinary(25))
print("The Binary of Decimal 0 is : "+DecimalToBinary(0))
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(number):
  product = 1
  while number != 0:
    #get the last digit from the number
    digit = number % 10
    #store the multiplied digits
    product *= digit
    #remove last digit from the number
    number = number // 10
  return product

print(prodDigits(345))
print(prodDigits(1234))
print(prodDigits(777))
```

9. Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:

  a. 1,2,3,4,5

  b. 12,13,14

```python
def multiply(*args):
  product = 1
  for i in args:
    #store the multiplied digits
    product *= i
  return product

print("Product of (1,2,3,4,5):", multiply(1, 2, 3, 4, 5))
print("Product of (12,13,14):", multiply(12, 13, 14))
```
